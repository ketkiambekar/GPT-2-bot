This code chooses a word from a sentence to consider as prefix for text generation by GPT-2

```
import string
import random


def choose_prefix(input_str,top_words):


  #Remove punctations
  input_str=input_str.translate(str.maketrans('', '', string.punctuation))
  top_longest=[]

  #Choose the top longest word in the sentence
  words = input_str.split()

  #top_words should not be longer than the sentence itself. 
  if top_words>len(words):
    top_words=1

  print(top_words)

  #remove two letter words
  for i in range(0,top_words):
    prefix = max(words, key=len)
    if len(prefix)>2:
      top_longest.append(prefix)
    words.remove(prefix)


  print(top_longest)
    
  #print(top_longest)
  if len(top_longest)==0:
    top_longest='Omg wtf No-way Maybe Perhaps However'.split()

  return random.choice(top_longest)



```

```
#Calling function

inp="Besides like a lot of women of that era she had her own creative and sometimes less than legal ways of making money from home and she utilized them all as the need arose."
top_words=4

inp ="She is"

print(choose_prefix(inp,top_words))
```

```
import re
import pandas as pd

file = 'prefix keywords.txt'
input_str ="Besides like a lot of women of that era she had her own creative and sometimes less than legal ways of making money from home and she utilized them all as the need arose."


def chose_prefix(file, input_str):

  file = open("prefix keywords.txt",'r+')
  content = file.read()
  keywords = content.split('\n')
  print(keywords)
  file.close()

  res = []
  for word in keywords:
    count = len(re.findall(rf'\b{word}\b',input_str))
    if count >0 and word is not '':
      res.append((word,count)) 
  print(res)

  df = pd.DataFrame(res,columns=["Word", "Count"])
  df.sort_values("Count", ascending = False,inplace=True)
  prefix = df['Word'].iloc[0]
  return prefix
```
